### We can also use the np.meshgrid function to create a relation from 1D arrray (array1) to another 1D array (array2)

### We can now plot a dot plot for the 2 matrices that we have created using the meshgrid function to understand how it works

### Lets plot something a little more interesting than a plain dot representation (sin function) 

```python
x = array1
y = array2
xx, yy = np.meshgrid(array1, array2)
z = np.sin(xx**2 + yy**2) / (xx**2 + yy**2)
h = plt.contourf(x,y,z)
```

### Oops. Since the plot looks very chaotic and "unappealing", lets plot the sin function of arrays having equal intervals as they must form better patterns

```python
x = np.arange(-10, 10, 1)
y = np.arange(-10, 10, 1)
xx, yy = np.meshgrid(x, y)
z = np.sin(xx**2 + yy**2) / (xx**2 + yy**2)
h = plt.contourf(x,y,z)
```

### Let's now see how different cmap values can change the above heatmap

# Abstract

### In the notebook we have used majorly 2 types of heat-maps, one of which is the plain heat map of a random matrix while the other is a heatmap of the sin function of the np.meshgrid of 2 randomly generated 1D arrays. The other heat-maps are just the same sin-function heatmaps but with different cmap values. These cmap values can be used for plotting different things to suit the purpose.

### The idea behind choosing a good colormap is to find a good representation in 3D colorspace for your data set. For many applications, a perceptually uniform colormap is the best choice --- one in which equal steps in data are perceived as equal steps in the color space.

### Colormaps are often split into several categories based on their function (see, e.g., [Moreland]):

*  Sequential: change in lightness and often saturation of color incrementally, often using a single hue; should be used for representing information that has ordering.  
*  Diverging: change in lightness and possibly saturation of two different colors that meet in the middle at an unsaturated color; should be used when the information being plotted has a critical middle value, such as topography or when the data deviates around zero.  
*  Cyclic: change in lightness of two different colors that meet in the middle and beginning/end at an unsaturated color; should be used for values that wrap around at the endpoints, such as phase angle, wind direction, or time of day.  
*  Qualitative: often are miscellaneous colors; should be used to represent information which does not have ordering or relationships.
